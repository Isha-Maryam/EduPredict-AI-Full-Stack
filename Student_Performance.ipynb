```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Expanded_data_with_more_features.csv')
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Drop unnamed column:-
df=df.drop("Unnamed: 0", axis=1)
print(df.head())
```

```python
#Change weekly study hours column:- Data transformation
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct","5-10")
print(df.head())
```

Gender Distribution:-

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df, x='Gender')
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

#From the above chart we have analysed that :
the number of females in the data is more than the number of males

```python
gb=df.groupby("ParentEduc").agg({"MathScore": 'mean', "ReadingScore" : 'mean', "WritingScore": 'mean'})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb, annot=True)
plt.title("Relation between Parent Education & Student's Score")
plt.show()
```

#From above chart its concluded that:
the education of parents have good impact on their child scores

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore": 'mean', "ReadingScore" : 'mean', "WritingScore": 'mean'})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1, annot=True)
plt.title("Relation between Parent Marital Status & Student's Score")
plt.show()
```

#From above chart its showing that:
 parents marital status
never impacting on childs scores

#Distribution of Ethnic Group:

```python
groupA= df.loc[(df["EthnicGroup"]=='group A')].count()
groupB= df.loc[(df["EthnicGroup"]=='group B')].count()
groupC= df.loc[(df["EthnicGroup"]=='group C')].count()
groupD= df.loc[(df["EthnicGroup"]=='group D')].count()
groupE= df.loc[(df["EthnicGroup"]=='group E')].count()
lbls = ['group A', 'group B', 'group C', 'group D', 'group E']
myList=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
print(myList)
plt.pie(myList, labels=lbls, autopct="%1.2f%%")
plt.title("Ethnic Group Distribution")
plt.show()
```
