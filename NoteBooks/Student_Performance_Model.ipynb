```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Expanded_data_with_more_features.csv')
print(df.head())
```

```python
df=df.drop("Unnamed: 0", axis=1)
print(df.head())
```

```python
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
  df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df['NrSiblings'].fillna(df['NrSiblings'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df['WklyStudyHours'].unique()
```

```python
df['WklyStudyHours']= df['WklyStudyHours'].map({
    '< 5':2,
    '5 - 10':7,
    '> 10':12
})
```

```python
df['WklyStudyHours'].head()
df.info()
```

```python
x=df.drop(columns=['MathScore'])
y=df['MathScore']
```

```python
x=pd.get_dummies(x,drop_first=True)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x,y, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
y_pred=model.predict(x_test)
print("R2 Score: " ,r2_score(y_test, y_pred))
print("MAE: ",mean_absolute_error(y_test, y_pred))
```
